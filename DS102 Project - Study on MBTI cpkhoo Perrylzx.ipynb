## Executive Summary

A brief literature review suggested that personality traits could affect word preferences. Thus, a simple text analysis was done on a forum posts to explore word preferences of each personality trait. The personality traits were based on Myers-Briggs Type Indicator (MBTI).

First, the data was treated to make them suitable for text analysis. After which, the top 30 words of each personality trait were identified. Target words were then chosen and the frequencies of the target words by each user were calculated. Z test was utilised to determine the statistical significance of each difference.

Results suggested that there were significant differences in word preferences. The differences were consistent with our hypotheses with the exception of the third hypothesis (T's use more objective words than F's). Limitations were then discussed prior to the conclusion.

## Research Topic & Hypotheses

The American Psychological Association (APA) defines personality as unique patterns of thoughts, emotions and behaviour of an individual. In attempts to understanding personality, many psychologists have developed many different theories. And these theories come with many different ways of measuring them. One such measure is the Myers-Briggs Type Indicator (MBTI). MBTI is a self-report questionnaire that identifies 16 unique personalities based on 4 axes:

- Introversion (I) – Extraversion (E) (Social interaction: I's enjoy being alone; E's prefer frequent social interaction)

- Intuition (N) – Sensing (S) (Information gathering: N's like to think about possibilities; S's pay more attention to current reality)

- Thinking (T) – Feeling (F) (Decision making: T's are more objective; F's are more emotional)

- Judging (J) – Perceiving (P) (Dealing with the world: J's prefer structure; P's are more flexible)

A person's personality type will be based on how they perform in each axis. 
Some examples of personality types are as below:

- INTJ (Introverted, Intuitive, Thinking, Judging)
- ESFP (Extroverted, Sensing, Feeling, Perceiving)
- ISFJ (Introverted, Sensing, Feeling, Judging)

You can take the free personality test here: https://www.16personalities.com/free-personality-test

<img src="https://upload.wikimedia.org/wikipedia/commons/8/81/Art-mb2-300x0.jpeg" />

MBTI was created by Katharine Cook Briggs and her daughter, Isabel Briggs Myers. Both mother and daughter were never formally educated in Psychology but were self taught. MBTI was developed to help women identify war-time jobs that would suit them best during World War 2. Since then, MBTI has been widely used by companies to filter suitable candidates and is also a very popular tool for people today to express their identity.

Personality types have been observed to display certain language preference. Hirsch and Peterson (2009) found that extroverted individuals tend to use words that are human related (e.g. people, family) more often than introverted people. In 2015, Farrokhi, Nourelahi and Elahi subjected eighty six Iranian EFL (English as Foreign Language) students to writing tests and concluded that ENFP's performed significantly better than other personality types in argumentative essay writings while INTP did the best in descriptive writings. These findings suggest distinct linguistic preferences between personality types.

With the brief literature review, we are able to come up with the following hypotheses:

**Hypothesis 1:**

Introversion (I) – Extraversion (E):
  
    H0: Extroverts do not use people related words more often than introverts.
    H1: Extroverts use people related words more often than introverts.

**Hypothesis 2:**

Intuition (N) – Sensing (S):
    
    H0: Individuals high in Intuition do not use counterfactual words more often than individuals high in Sensing.
    H1: Individuals high in Intuition use counterfactual words more often than individuals high in Sensing.

**Hypothesis 3:**

Thinking (T) – Feeling (F):

    H0: Individuals high in Thinking do not use objective words more often than individuals high in Feeling.
    H1: Individuals high in Thinking use objective words more often than individuals high in Feeling.

**Hypothesis 4:**

Judging (J) – Perceiving (P):

    H0: People high in Judging do not use rigid words more often than people high in Perceiving.
    H1: People high in Judging use rigid words more often than people high in Perceiving.


We will be using the Z test to test for statistical significance:


$$
Z  = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$


Alpha level is set at .05 and the tests are one directional. Thus, Z value will have to be 1.64 and above for the difference to be significant.

### Data Cleaning

Please note that "posts" was left in the dataframe for comparison with cleaned data.

```python
#### Defining function to clean data
eng_stopwords = stopwords.words('english')
df_to_clean = df.copy()

def clean_data(words):    
    temp_list = ''
    words = words.lower()
    words = re.sub(r'http\S+', '', words) #Removes URL's
    words = re.sub("-", "", words) # Accounts for words joined by '-'
    words = re.sub("[^a-zA-Z]", " ", words) # Keep words only
    words = re.sub(' +', ' ', words) # Removes >1 spaces 
    words = re.sub('sent from my \w+ using tapatalk', '', words) # Removes signatures
    words = re.sub('sent from my \w+', '', words) # Removes cut signatures
    words = word_tokenize(words)
    for word in words:
        if word not in eng_stopwords:
            temp_list+=(lemmatizer.lemmatize(word) + ' ') 
    return temp_list

df_to_clean['cleaned'] = df_to_clean['posts'].apply(clean_data)
display(df_to_clean.head())
df_to_clean.to_csv('mbti_cleaned_lem.csv', index=False)
```

### Z Test (I vs E) - Calculating People Related Words per Post

```python
ei_target = e_target

for word in i_target:
    if word not in e_target:
        ei_target.append(word)

def people_count(posts):
    temp_count = 0
    posts = posts.split()
    for word in posts:
        if word in ei_target:
            temp_count += 1
    return temp_count

e_df['people words count'] = e_df['cleaned'].apply(people_count)
i_df['people words count'] = i_df['cleaned'].apply(people_count)

display(e_df.head(10))
display(i_df.head(10))
```

### Z Test (I vs E) - Distributions of People Related Words

```python
#### Distribution Plot ####

fig = plt.figure(figsize = (16,9))
ax1 = fig.add_subplot(211)
sns.distplot(e_df['people words count'], ax = ax1, color = 'orange')
plt.title('Distribution of People Related Words in Top 30 Words Used by Extroverts', fontsize = 14)
plt.ylabel('Frequency', fontsize = 12)
plt.xlabel('People Related Words', fontsize = 12)


ax2 = fig.add_subplot(212)
sns.distplot(i_df['people words count'], ax = ax2, color = 'orange')
plt.title('Distribution of People Related Words in Top 30 Words Used by Introverts', fontsize = 14)
plt.ylabel('Frequency', fontsize = 12)
plt.xlabel('People Related Words', fontsize = 12)
plt.subplots_adjust(top = 1.2)
plt.show()
```

### Z Test (I vs E) - Calculating the Z Value

```python
e_mean = e_df['people words count'].mean()
e_std = e_df['people words count'].std()
e_sample = len(e_df['people words count'])
i_mean = i_df['people words count'].mean()

z_test_ei = round((e_mean - i_mean) / (e_std / math.sqrt(e_sample)),2)

results_ei = ''

if z_test_ei >= 1.96:
    results_ei = ', indicating that the difference is significant. The null hypothesis is rejected.'
elif z_test_ei <= 1.96:
    results_ei = ', indicating that the difference is significant but in another direction. The null hypothesis is accepted.'
else:
    results_ei = ', indicating that the difference is not significant and could be due to chance. The null hypothesis is accepted.'

print('The Z value calculated is ' + str(z_test_ei) +  results_ei)
```

### Z Test (T vs F) - Calculating Objective Words per Post

```python
tf_target = t_target

for word in f_target:
    if word not in t_target:
        tf_target.append(word)

def objective_count(posts):
    temp_count = 0
    posts = posts.split()
    for word in posts:
        if word in tf_target:
            temp_count += 1
    return temp_count

t_df['objective words count'] = t_df['cleaned'].apply(objective_count)
f_df['objective words count'] = f_df['cleaned'].apply(objective_count)

display(t_df.head(10))
display(f_df.head(10))
```

### Z Test (T vs F) - Calculating the Z Value

```python
t_mean = t_df['objective words count'].mean()
t_std = t_df['objective words count'].std()
t_sample = len(t_df['objective words count'])
f_mean = f_df['objective words count'].mean()

z_test_tf = round((t_mean - f_mean) / (t_std / math.sqrt(t_sample)),2)

results_tf = ''

if z_test_tf >= 1.96:
    results_tf = ', indicating that the difference is significant. The null hypothesis is rejected.'
elif z_test_tf <= 1.96:
    results_tf = ', indicating that the difference is significant but in another direction. The null hypothesis is accepted.'
else:
    results_tf = ', indicating that the difference is not significant and could be due to chance. The null hypothesis is accepted.'
    
print('The Z value calculated is ' + str(z_test_tf) +  results_tf)
```

## Insights and Evaluation 

- Insight 1 - Extroverts use more people related words than Introverts. H0 rejected.
- Insight 2 - Individuals high in Intuition use more counterfactual words than individuals high in Sensing. H0 rejected.
- Insight 3 - Individuals high in Feeling use more objective words than high Thinking individuals. H0 not rejected.
- Insight 4 - Individuals high in Judging use more rigid words than individuals high in Perceving. H0 rejected.

All null hypotheses were rejected with the exception of the 3rd hypothesis (T's use more objective words than F's). Contrary to our hypothesis, individual high in Feeling in our sample used more objective words than individuals high in Thinking (Z = -10.79, p < .05).

Of the three accepted alternative hypotheses, the effects of the Introversion-Extraversion axis and the Intuition-Sensing axis were very strong with the Z scores of 20.03 (p < .05) and 11.98 (p < .05) respectively. The effect of the Judging-Perceiving axis was significant but not as strong with the Z score of 3.31 (p < .05).

### Limitations

Like all studies, we were met with several limitations. Firstly, the sample was not evenly distributed, this can be seen in the graph below.
Secondly, possible interaction effects (e.g.ENTJ might use more people words than ESTJ) were not accounted for as each sample was only grouped into one personality trait at one time.
Furthermore, the target words were selected based on our own discretion and only from the top 30 words used by each group. Results would vary greatly depending on the words selected.
Finally, the context of each document was not taken into account. Words could have different meanings when used differently.
Future studies should take these limitations into consideration for better results.
